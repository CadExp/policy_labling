```python
import openpyxl

book = openpyxl.load_workbook('21275.xlsx')

sheet = book.active
```

```python
data = []
for row in sheet.iter_rows(min_row=1, min_col=1, max_row=sheet.max_row, max_col=sheet.max_column):
    row_data= [cell.value for cell in row]
    data.append(row_data)
len(data)
```

# 政策标签标注

```python
policy_labels = [
    "财税扶持",
    "资质认定",
    "资金补贴",
    "人才激励",
    "课题项目",
    "政府采购",
    "赛事活动",
]
```

```python
import json
with open("policy_labels.json") as f:
    label_json = json.load(f)
label_json
```

```python
label_data = label_json["data"]
label_name_list = [i["name"] for i in label_data]
label_name_list
```

```python
def get_label(frame_row):
    if frame_row["nature"] in label_name_list:
        return frame_row["nature"]
    elif frame_row["type"] in label_name_list:
        return frame_row["type"]
    elif frame_row["big_class"] in label_name_list:
        return frame_row["big_class"]
    return None
frame["new_label"] = None
for i in range(len(frame)):
    row_location_data = frame.iloc[i]
    row_location_data["new_label"] = get_label(row_location_data)
frame[["nature", "type", "big_class", "new_label"]]
```

```python
frame[frame["new_label"].isnull()]
```

```python
from chean_data import cleaning_data
# 清洗content
for i in range(len(frame)):
    row_location_data = frame.iloc[i]
    row_location_data["content"] = cleaning_data(row_location_data["content"])
```

```python
frame[frame["new_label"].isnull()][["title", "content"]]
```

```python
from openpyxl import Workbook
new_book = Workbook()
new_sheet = new_book.active
cols_names = list(frame.columns)
new_sheet.append(cols_names)
for i in range(len(frame)):
    sheet_row = [frame.iloc[i][name] for name in cols_names]
    new_sheet.append(sheet_row)
new_book.save("sample.xlsx")
```
